```python
import pandas as pd
df = pd.read_csv('items.csv')
df = df.eval("importe = precio * cantidad")
dfa = df.query("categoria == 'A'")
promedios_por_categoria = df.groupby('categoria')['importe'].mean()
promedio_A = promedios_por_categoria['A']
resp = round(promedio_A, 3)
print(resp)
```

```python
dfc = pd.read_csv('clientes.csv')
dfo = pd.read_csv('ordenes.csv')
df_merged = pd.merge(dfc, dfo, on='cliente_id', how='left')
monto_por_cliente = df_merged.groupby('cliente_id')['monto'].sum().fillna(0)
respuesta_final = monto_por_cliente.get(1, 0)
print("Cliente:", respuesta_final)
```

```python
df = pd.read_csv('sensores.csv', usecols=['ts', 'temperatura', 'humedad'])
df['ts'] = pd.to_datetime(df['ts'], format='mixed', errors='coerce')
mediana_original = df['temperatura'].median()
media_original = df['humedad'].mean()
df['temperatura'] = df['temperatura'].fillna(mediana_original)
df['humedad'] = df['humedad'].fillna(media_original)
respuesta_a = df['ts'].isna().sum()
respuesta_b = round(df['temperatura'].median(), 1)
respuesta_c = round(df['humedad'].mean(), 3)
print(f"a) {respuesta_a}")
print(f"b) {respuesta_b}")
print(f"c) {respuesta_c}")
```
