#### Conceptos

La minería de datos es un conjunto de técnicas y procesos que permiten extraer patrones significativos, tendencias y información útil a partir de grandes volúmenes de datos. Esta práctica se puede considerar un subcampo de la inteligencia artificial y la estadística, y su objetivo principal es transformar datos sin procesar en información comprensible y aplicable.

Ciencia de datos

La ciencia de datos es un campo que combina las matemáticas y la estadística, la programación especializada, los analytics avanzados, la inteligencia artificial IA y el machine learning con conocimientos específicos en la materia para descubrir insights procesables ocultos en los datos de una organización.

IPython Magic Commands

```python
#Sirve para evaluar el desempeño de una expresion en cuanto a tiempo
In [8]: %timeit L = [n ** 2 for n in range(1000)]
```

```python
# Con doble porcentaje podemos evaluar el desempeño en cuanto a tiempo
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```
